### json chunking

```python
import json

with open(file[0], 'w', encoding='utf-8') as f:
    data = json.load(f)
```

```python
# 청킹 함수
def chunk_data(data, chunk_size=500):
    chunks = []
    for entry in data:
        page = entry["page"]
        content = entry["md"]
        
        # 텍스트를 chunk_size 기준으로 나누기
        while len(content) > chunk_size:
            split_index = content[:chunk_size].rfind("\n")  # 마지막 줄바꿈 기준으로 자름
            if split_index == -1:  # 줄바꿈이 없으면 그냥 chunk_size 기준으로 자름
                split_index = chunk_size
            chunks.append({
                "page": page,
                "chunk": content[:split_index].strip()
            })
            content = content[split_index:].strip()
        
        # 남은 내용 추가
        if content:
            chunks.append({
                "page": page,
                "chunk": content
            })
    return chunks

# 실행
chunked_data = chunk_data(data, chunk_size=100)

# 결과 출력
json.dumps(chunked_data, ensure_ascii=False, indent=4)
```

```python
def chunk_data_with_pages(data, chunk_size=500):
    chunks = []
    current_chunk = ""
    current_pages = []

    for entry in data:
        page = entry["page"]
        content = entry["md"]

        # 페이지가 바뀔 때마다 처리
        if page not in current_pages:
            current_pages.append(page)

        while len(content) > 0:
            if len(current_chunk) + len(content) <= chunk_size:
                current_chunk += content
                content = ""
            else:
                split_index = chunk_size - len(current_chunk)
                current_chunk += content[:split_index].strip()
                content = content[split_index:].strip()
            
            # 청크가 가득 차면 저장
            if len(current_chunk) >= chunk_size:
                chunks.append({
                    "pages": current_pages,
                    "chunk": current_chunk.strip()
                })
                current_chunk = ""
                current_pages = [page]

    # 남아 있는 내용 저장
    if current_chunk:
        chunks.append({
            "pages": current_pages,
            "chunk": current_chunk.strip()
        })
    
    return chunks

# 실행
chunked_data = chunk_data_with_pages(data, chunk_size=100)

# 결과 출력
json.dumps(chunked_data, ensure_ascii=False, indent=4)
```
